```python
from sklearn.cluster import AgglomerativeClustering

def hierarchical_clustering(X):
    hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
    return hc.fit_predict(X)

hc_data = hierarchical_clustering(data.T)
hc_data.shape
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, SpectralClustering
from sklearn.decomposition import PCA
def compare_clustering(X):
    print("K-means clustering")
    kmeans = KMeans(n_clusters=5, random_state=0).fit(X)
    kmeans_labels = kmeans.labels_
    print(kmeans_labels.shape)
    print('Hierarchical clustering')
    hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
    hc_labels = hc.fit_predict(X)

    print('DBSCAN clustering')
    dbscan = DBSCAN(eps=3, min_samples=2).fit(X)
    dbscan_labels = dbscan.labels_

    print('Spectral clustering')
    sc = SpectralClustering(n_clusters=5, affinity='nearest_neighbors', assign_labels='kmeans')
    sc_labels = sc.fit_predict(X)

    print('PCA for visualization')
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)

    # Plotting the results
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))
    axs[0, 0].scatter(x, y, c=kmeans_labels,cmap='inferno')
    axs[0, 0].set_title('K-means clustering')
    axs[0, 1].scatter(x, y, c=hc_labels,cmap='inferno')
    axs[0, 1].set_title('Hierarchical clustering')
    axs[1, 0].scatter(x, y, c=dbscan_labels,cmap='inferno')
    axs[1, 0].set_title('DBSCAN clustering')
    axs[1, 1].scatter(x, y, c=sc_labels,cmap='inferno')
    axs[1, 1].set_title('Spectral clustering')
    
    plt.show()
```

```python
pca.shape
```

```python
compare_clustering(pca)
```
